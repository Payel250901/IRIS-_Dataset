**Line Plot :**
Plot a line graph showing the sepal length of the three species of Iris flowers. Use
different colors to represent each species.

```python

setosa_data = data[data["Species"] == "Iris-setosa"]
versicolor_data = data[data["Species"] == "Iris-versicolor"]
virginica_data = data[data["Species"] == "Iris-virginica"]

setosa_sepal_length = setosa_data["SepalLengthCm"]
versicolor_sepal_length = versicolor_data["SepalLengthCm"]
virginica_sepal_length = virginica_data["SepalLengthCm"]
```

```python
plt.plot(setosa_data["SepalLengthCm"], label="Iris setosa", color="blue")
plt.plot(versicolor_data["SepalLengthCm"], label="Iris versicolor", color="green")
plt.plot(virginica_data["SepalLengthCm"], label="Iris virginica", color="red")

plt.xlabel("Sample Number")
plt.ylabel("Sepal Length (cm)")
plt.title("Sepal Length of Iris Flowers by Species")

plt.legend()

plt.show()
```

**Scatter Plot :** Create a scatter plot showing the relationship between the petal length and petal
width. Use different colors to represent each species.

```python
import seaborn as sns
```

```python
sns.scatterplot(x="PetalLengthCm", y="PetalWidthCm", data=data,hue="Species")

plt.xlabel("PetalLength(Cm)")
plt.ylabel("PetalWidth(Cm)")
plt.title("Petal Length vs. Petal Width for Iris Flowers")


plt.grid(True)

plt.show()
```

**Histogram :**
 Plot a histogram showing the distribution of the petal widths. Use different colors to
represent each species and display a legend.

```python
plt.figure(figsize=(10, 6))
plt.hist(
    [data[data["Species"] == "Iris-setosa"]["PetalWidthCm"], data[data["Species"] == "Iris-versicolor"]["PetalWidthCm"], data[data["Species"] == "Iris-virginica"]["PetalWidthCm"]],
    bins=10,
    histtype="bar",
    stacked=False,
    label=["Iris-setosa", "Iris-versicolor", "Iris-virginica"],
    rwidth=0.8,
    color=["blue", "green", "red"],
)

# Add labels and title
plt.xlabel("Petal Width (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Widths for Iris Species")

# Add legend
plt.legend()
```

**Pie Chart :** Assume that the dataset represents the entire population of Iris flowers. Create a pie
chart showing the proportion of each species in the population.

```python
species_counts = data["Species"].value_counts()

pie_chart_slice_labels = species_counts.index
pie_chart_slice_values = species_counts.values

# Create the pie chart
plt.pie(pie_chart_slice_values, labels=pie_chart_slice_labels, autopct="%1.1f%%")
plt.title("Proportion of Iris Species in the Population")

# Legend is not applicable to pie charts, comment it out
plt.legend(title='Species Type Distribution', bbox_to_anchor=(0.8, 0, 0.5, 1), fontsize='12')

plt.show()
```
